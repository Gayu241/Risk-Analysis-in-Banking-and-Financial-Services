# Handeling Outliers

```python
data.describe()
```

```python
column = 'CNT_CHILDREN'
Q1 = data[column].quantile(0.25)
Q3 = data[column].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data1 = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
print(data1)
```

```python
data1.CNT_CHILDREN.describe()
```

```python
columns_to_check = ['DAYS_EMPLOYED', 'CNT_FAM_MEMBERS','OBS_60_CNT_SOCIAL_CIRCLE','DEF_60_CNT_SOCIAL_CIRCLE']  # Replace with your columns
Q1 = data1[columns_to_check].quantile(0.25)
Q3 = data1[columns_to_check].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data2= data1[~((data1[columns_to_check] < lower_bound) | (data1[columns_to_check] > upper_bound)).any(axis=1)]
print(data2)
```

```python
data2.describe()
```

```python
column_name = 'DAYS_EMPLOYED'
data2[column_name] = data2[column_name].abs()
print(data2.head())
```
